### Mascotas

```python
#Objeto Mascota
class Mascota(object):
  #Constructor
    def __init__(self):
      self._nombre = None
      self._hambre = 0
      self._salud = 0
      self._higiene = 0
      self._edad = 0
      self._dias = [0,0,0] # 0 = dias con hambre, 1 = dias con mala salud, 2 = dias cochino

  #Destructor
    def __del__(self):
        pass

    #Regresa el nombre
    def getNombre(self):
      return self._nombre
    #Regresa la salud
    def getSalud(self):
      return self._salud
    #Regresa la edad
    def getEdad(self):
      return self._edad
    #Regresa los dias
    def getDias(self):
      return self._dias

    #Cambia la salud
    def setSalud(self, e):
      self._salud = e
    #Cambia nombre
    def setNombre(self):
      print("Nuevo nombre para ", self._nombre, ": ")
      self._nombre = input()

    #actualiza el arreglo de dias
    def actualizaDias(self):
        stats = [self._hambre, self._salud, self._higiene]
        for x in range(len(stats)):
          if stats[x] == 1:
            self._dias[x] += 1

        for x in range(len(self._dias)):
          if x == 0:
            print("   [Dias hambriento: ",self._dias[x], "]")
          if x == 1:
            print("   [Dias con mala salud: ",self._dias[x],"]")
          if x == 2:
            print("   [Dias sucio: ",self._dias[x],"]")

    #Imprime datos (polimorfismo)
    def imprimeDatos(self):
        print(" Estado de ", self._nombre, ":")
        print("   Salud: ", self._salud)
        print("   Hambre: ", self._hambre)
        print("   Higiene: ", self._higiene)
        print("   Edad: ", self._edad)

    #Verifica el estado de la mascota
    def Estado(self):

        if self._edad > 5 :
          print("   Tu mascota murio de edad")
          return False
        if self._dias[0] > 2 :
          print("   Tu mascota murio de hambre")
          return False
        if self._dias[1] > 2 :
          print("   Tu mascota murio por la enfermedad que no curaste")
          return False
        if self._dias[2] > 2 :
          self._salud = 1
          print("   Tu mascota se enfermo de lo sucio que esta")
          return True
        else:
          print("   Tu mascota sigue viva")
          return True

    #Realiza los eventos aleatorios (polimorfismo)
    def Eventos(self, e):

      print("   ", self._nombre, "esta hambriento.")
      self._hambre = 1
      print("   ",self._nombre, "tiene ", self._edad, "dias de edad")
      self._edad += 1
      print("   -->Evento del dia para el ", self.getTipo(), " ", self._nombre)

      if e == 1 :
        self._salud = 1
        print("   ",self._nombre,"Se enfermo, necesita ir al veterinario ")
      if e == 3:
        print("   Nada")

#Objeto Gato
class Gato(Mascota):

  #Constructor
    def __init__(self):
      super().__init__()
      self._nombre = input("Ingrese nombre de su gato: ")
      self.__tipo = "gato"
      self.arenero = None
      print("Se ha creado un ", self.__tipo)

  #Destructor
    def __del__(self):
      pass

    #Regresa el tipo
    def getTipo(self):
      return self.__tipo

    #Imprime datos (polimorfismo)
    def imprimeDatos(self):
      super().imprimeDatos()
      print("   Arenero: ", "Si" if self.arenero != None else "No")
      if self.arenero != None:
        print("   Con arena " if self.arenero.getRelleno() else "   Sin arena")

    #Realiza eventos (polimorfismo)
    def Eventos(self, e, p):
      super().Eventos(e)
      if e == 2 :
        print("   Necesita hacer sus necesidades (verifica la arena)")
        if self.arenero == None:
          self._higiene = 1
        else:
          self.arenero.ar = None

#Objeto perro
class Perro(Mascota):

  #Constructor
    def __init__(self):
      super().__init__()
      self._nombre = input("Ingrese nombre de su perro: ")
      self.__tipo = "perro"
      self.popo = False
      print("Se ha creado un ", self.__tipo)

  #Destructor
    def __del__(self):
        pass

    #Regresa el tipo
    def getTipo(self):
      return self.__tipo

    #Imprime datos (polimorfismo)
    def imprimeDatos(self):
        super().imprimeDatos()
        print("   Necesidades: ", "quiere hacer popo" if self.popo  else " todo bien")

    #Realiza eventos (polimorfismo)
    def Eventos(self, e, p):
      super().Eventos(e)
      if e == 2 :
        if self.popo :
          if p :
            print("   Salio a hacer popo.")
            self.popo = False
          else :
            print("   Se hizo popo encima porque no abriste la puerta")
            self._higiene = 1
            self.popo = False
        else:
          self.popo = True
          print("   Necesita hacer sus necesidades (abre la puerta)")
```

### Inventario/Tienda

```python
#Clases para objetos del inventario/tienda
class Objeto(object):
  def __init__(self):
    self._costo = None
    self._tipo = None
  def __del__(self):
    pass
  def getTipo(self):
    return self._tipo
  def getCosto(self):
    return self._costo

class comidaPerro(Objeto):
  def __init__(self):
    self._costo = 30
    self._tipo = "croquetas para perro"
  def __del__(self):
    pass

class comidaGato(Objeto):
  def __init__(self):
    self._costo = 20
    self._tipo = "croquetas para gato"
  def __del__(self):
    pass

class shampoo(Objeto):
  def __init__(self):
    self._costo = 30
    self._tipo = "shampoo"
  def __del__(self):
    pass

class arena(Objeto):
  def __init__(self):
    self._costo = 10
    self._tipo = "arena"
  def __del__(self):
    pass

class arenero(Objeto):
  def __init__(self):
    self._costo = 100
    self._tipo = "arenero"
    self.ar = False
  def __del__(self):
    pass
  def getRelleno(self) :
    return self.ar
  def setRelleno(self, e) :
    self.ar = e
```
